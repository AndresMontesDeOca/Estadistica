## Distribuciones

```python
# Discretas: Bernulli, Binomial, Poisson

# Bernoulli:
# f(x) = p^x*q^(1-x) # x={0, 1})
# V = p*q


# Binomial: "x cantidad de Exitos en n repeticiones"
# "La Binomial es una suma de Bernulli independientes"
# f(x) = (n/x)*p^x*q^(n-x)
# (n/x) = np.divide(n!, x!*(n-x)!)

# Poisson: " Cantidad de ocurrencias de un Evento sobre un continuo"
# Continuas: Normal
# z = np.divide(x-u, d)
```

### Funciones Auxiliares

```python
# Funciones axuliares que ayudan a la resolucion de los ejericios de la guia

# Calculo de la Esperanza
def calcular_esperanza(data_):
    esperanza = data_.Xi_fi.sum()
    return np.float(esperanza)

# Calculo el desvio de la variable
def calcular_desvio(data_):
    # Inicializamos
    varianza_serie = pd.Series(np.float64) # inicializamos la serie
    # La vamos cargando de a registro
    varianza_serie = data_.apply(lambda x: np.square(x.Xi)*x.fi, axis=1)
    result = varianza_serie.sum() - np.square(calcular_esperanza(data_))
    return np.sqrt(result)

# Funcion que calcula el n
def calcular_n(data_):
    result = data_.fi.sum()
    return result

# Funcion que genera Tabla de Frecuencias n, y la media: Recibe DataFrame con Xi, fi
def generar_tabla_frecuencias(data_):
    # Primero genero Xi * fi
#     display(data_)
    Xi_fi = data_.Xi * data_.fi
    Xi_fi.name = 'Xi_fi'
    
    # Genero frecuencias acumuladas
    x_ant = 0
    Fi = []
    for x in data_.fi:
        Fi.append(x+x_ant)
        x_ant += x
    Fi = pd.Series(Fi, name='Fi')
    
    # Genero Serie Frecuencia Relativa y Frecuencia Acumulada
    fri = data_.fi.apply(lambda x: np.divide(x, calcular_n(data_)))
    fri = pd.Series(fri, name='fri')
    Fri = Fi.apply(lambda x: np.divide(x, calcular_n(data_)))
    Fri = pd.Series(Fri, name='Fri')
    
    # Concateno y devuelvo
    result = pd.concat([data_,Xi_fi, Fi, fri, Fri], axis=1)
    return result

def generar_binomial(p_, n_):
    q = 1 - p_
    n_values = range(0, n_+1)
    n_fact = np.math.factorial(n_)
    P_serie = []
    
    # Recorro para todo x en n
    for x in n_values:
        x_fact = np.math.factorial(x) # Calculo el X factorial para esta vuelta
        n_x = np.divide(n_fact, x_fact*np.math.factorial(n_-x)) # Calculo el combinatorio
        P = n_x*p_**x*q**(n_-x) # Calculo la probabilidad puntual en x
        P_serie.append(P) # appendo
    
    # Le doy forma y devuelvo
    P_serie = pd.Series(P_serie, name='fi')
    x_serie = pd.Series(n_values, name='Xi')    
    result = pd.DataFrame(pd.concat([x_serie, P_serie], axis=1))
    return pd.DataFrame(result)

# data = generar_tabla_frecuencias(data)
# data
```

```python
# Ejercicio modelo en clase
header_serie = pd.Series(np.arange(4), name='Xi')
values_serie = pd.Series([Fraction(1, 8), Fraction(3, 8), Fraction(3, 8), Fraction(1, 8)], name='fi')

data = pd.concat([header_serie, values_serie], axis=1)
data = generar_tabla_frecuencias(data)
display(data)
print('Esperanza:', calcular_esperanza(data))
print('Desvio:', calcular_desvio(data))
```
