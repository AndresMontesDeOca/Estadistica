```python
u = 1780
x = 1900
d = 110
n = 40
alpha = .05

Ze = np.divide(x-u, 110/np.sqrt(n))
print('Z estimado:', Ze)
print('Z observado:', st.norm.ppf(.95))
print('P value:', st.norm.cdf(6.8995)-1)
```

```python
# Usando SciPy
population = np.random.normal(loc=u, scale=d, size=9999999) # N infinito, utilizo valor grande
sample = np.random.normal(loc=x, size=n) # Seteo el desvio en la population

st.ttest_ind(a=population, b=sample, equal_var=True, alternative='less')
```

```python
u = 15
x = 12.2
n = 50
d = 6.2
alpha = .05
Tobs = np.divide(x-u, d/np.sqrt(n))
Tcrit = st.t.ppf(alpha, n-1)
print('T Critico:', Tcrit)
print('T Observado:', Tobs)
```
